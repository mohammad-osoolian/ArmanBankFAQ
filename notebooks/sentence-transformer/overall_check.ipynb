# imports

```python
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer
```

```python
sheet_path = '/content/Ka-ChatBot_BenchMark.xlsx'
models = ['PartAI/Tooka-SBERT-V2-Large', 'intfloat/multilingual-e5-base', 'google/embeddinggemma-300m', 'xmanii/maux-gte-persian', 'intfloat/multilingual-e5-base', 'sentence-transformers/LaBSE', 'BAAI/bge-m3']
```

# load and prepare data

```python

faq_df = pd.read_excel(sheet_path, sheet_name="faq")
samples_df = pd.read_excel(sheet_path, sheet_name="samples")
```

```python
faq = faq_df['faq'].tolist()
samples = samples_df['sample'].tolist()
labels = torch.tensor(samples_df['gt_idx'].tolist())
```

# experiment candidate models



```python
precisions = {}
model = ''
for modelname in models:
  del model
  torch.cuda.empty_cache()
  print(f"#### Experimenting on Model: {modelname} ####")
  model = SentenceTransformer(modelname, trust_remote_code=True)
  faq_emb = model.encode(faq)
  samples_emb = model.encode(samples)
  similarities = model.similarity(faq_emb, samples_emb)

  predict_indices = torch.max(similarities, axis=0)[1]
  predicts = predict_indices + 1

  acc = (predicts == labels).sum() / labels.shape[0]
  precisions[modelname] = acc.clone().item()
```

# results

```python
items = sorted(precisions.items(), key=lambda x: x[1], reverse=True)
for item in items:
  print(f"{item[0]}: {item[1]}")
```
